Zeige die Gültigkeit der Ungleichung $n^2 - 4n + 3 > 0$ für alle natürlichen Zahlen $n \geq 5$.

IA: n = 5 $\to$ $5^2 - 4 \cdot 5 + 3 = 25 - 20 + 3 = 8 > 0$ ($\checkmark$)

IV: $n^2 - 4n + 3 > 0$ für ein beliebiges $n \geq 5$.

IS: $n \to n + 1$<br>
$(n + 1)^2 - 4(n + 1) + 3$<br>
$= n^2 + 2n + 1 - 4n - 4 + 3$<br>
$= n^2 - 4n + 3 + 2n -3$<br>
$>$ (IV) $0 + 2n - 3$<br>
$= 2n - 3$<br>
$> 0$ ($\checkmark$)
