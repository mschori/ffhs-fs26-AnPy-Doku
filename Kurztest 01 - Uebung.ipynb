Zeige die GĂĽltigkeit der Ungleichung $n^2 - 4n + 3 > 0$ fĂĽr alle natĂĽrlichen Zahlen $n \geq 5$.

IA: FĂĽr $n=5$ gilt $5^2 - 4\cdot 5 + 3 = 25 - 20 + 3 = 8 > 0$  ($\checkmark$)

IV: $n^2 - 4n + 3 > 0$ fĂĽr ein beliebiges $n \geq 5$.

IS: <br>
$n \to n+1$<br>
$n^2 - 4n + 3 > 0 \Rightarrow (n+1)^2 - 4(n+1) + 3 > 0$<br>
$= n^2 + 2n + 1 - 4n - 4 + 3$<br>
$= n^2 - 2n$<br>

$\to$ Differenz zu IV berechnen:<br>
$(n^2 - 2n) - (n^2 - 4n + 3))$<br>
$= n^2 - 2n - n^2 + 4n - 3$<br>
$= 2n - 3$<br>

$\to$ Beweisen, dass $n^2 - 2n = (n^2 - 4n + 3) + (2n - 3)$<br>
$n^2 - 2n = n^2 - 4n + 3 + 2n - 3$<br>
$n^2 - 2n = n^2 - 2n$  ($\checkmark$)
