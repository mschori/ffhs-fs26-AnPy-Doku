# Konvergenz von Folgen
=> Ist wichtig für Prüfung und auch Abschlussprüfung.

Definition: Sei a_m eine Folge. Falls der Limis, lim a_m = g (endlich) ist, dann ist a_m konvergent.

g = Grenzwert

Monotonie fallend: a_m < a_{m+1}<br>
Monotonie steigend: a_m > a_{m+1}<br>
Monotonie bedeutet: Die Werte werden immer kleiner oder immer grösser. => es gibt keine Abwechslung. Dabei dürfen die Werte auch jeweils <= (steigend) oder >= (fallend) sein. Wenn aber < oder > ist, gilt die Bezeichnung (streng monoton).

1) Bildliche Darstellung als Graph

Die Ergebnisse nähern sich immer mehr dem Grenzwert (g) an. Wir können ein Epsilon definieren, das über und unter dem Grenzwert liegt. Wenn die Punkte dann nicht den "Schlauch" übertreten, dann ist die Folge konvergent. Wenn sie aus dem Schlauch austritt, ist sie divergent.

## Kernaussage
- Satz: Monoton + Beschränkt = Konvergenz
- ABER: Aus der Konvergenz folgt nicht zwingend die Monotonie
- Satz: Konvergenz => Beschränktheit
- - Viele obere Schranken und viele untere Schranken sind möglich
- - Nur die kleinste obere Schranke und die grösste untere Schranke sind dabei interessant


# 0-Folge
Das ist eine Folge die den Grenzwert 0 hat. a_m mit lim a_m = 0 = g.

Beispiel: $\frac{1}{m}$ ist eine Nullfolge.

Beispiel: $(\frac{1}{2})^m = \frac{1}{2^m}$ ist ebenfalls eine Nullfolge.

Beispiel: $c_m = q^m$ mit $|q| < 1$ ist eine Nullfolge.

Beispiel: $d_m = \frac{m}{m^2 - 1} = \frac{m}{m^2 (1 - \frac{1}{m^2})} = \frac{1}{m} * \frac{1}{1 - \frac{1}{m^2}}$ ist auch eine Nullfolge.

Beispiel: $e_m = 2^m = 2^0, 2^1, 2^2, ...$ ist keine Nullform, da sie immer weiter steigt.

# Mastertheorem
Beispiel: Ich habe ein Problem und teile dieses in drei kleinere Probleme auf => Divide-and-Conquer

Das Mastertheorem ist ein Werkzeug, um Rekursionen irgendwie berechnen zu können. Oder eine Aussage über Rekursionen treffen zu können.

---

Mastertheorem vereinfacht dargestellt: $T(m) = b * T(\frac{m}{c}) + f(m)$

=> Wir teilen in b Teilprobleme auf (also Anzahl der Teilprobleme)<br>
=> Jedes Teilproblem hat die Grösse $\frac{m}{c}$<br>
=> f(m): sind die Kosten für das Aufteilen und Mergen

---

### Fall 1: Rekursion dominiert
Beispiel: wir teilen Problem immer in zwei Teile auf (also b = 2). Plus wir machen fast keine Zusatzarbeit (also fast kein f(m)).

### Fall 2: Balance
Rekursion und Zusatzarbeit sind etwa gleich wichtig (kosten ähnlich viel).

So verhalten sich oft effiziente Algorithmen.

## Formulierung des Mastertheorems (einfache Form)
Eine Funktion $T(n)$ erfülle folgende Rekursionsgleichung.

$$
T(n) =
\begin{cases}
a & \text{falls } n \le 1 \\
d \cdot T(n/b) + cn & \text{sonst}
\end{cases}
$$

Dann gilt:

- $T(n) \in O(n)$ falls $d < b$
- $T(n) \in O(n * log(n))$ falls $d = b$
- $T(n) \in O(n^{log_b(d)})$ falls $d > b$

## Allgemeine Form des Mastertheorems
Die einfache Form ist nicht sehr nützlich. Für das Merge-Sort Beispiel kann es nicht direkt angewendet werden. Das liegt daran, dass die Rekursion beim Merge-Sort um den Term $\frac{n}{2}$ noch Balken hat. Also $|\frac{n}{2}|$.

Anstelle des Quotienten $\frac{d}{b}$ möchte man den aufgerundeten Quotienten benützen dürfen; ebenso möchte man in gewissen Anwendungen zulassen, dass im Rekurs an der Stelle von $T(\frac{d}{b})$ der etwas grössere Term $T(\frac{d}{b} + e)$ verwendet werden darf.

Das führt zu folgender Version des Mastertheorems:

Die Funktion $T(n)$ erfüllt die folgenden Ungleichungen.

$$
T(n) =
\begin{cases}
a & \text{falls } n \le n_0 \\
d \cdot T([n/b]) + cn & \text{sonst}
\end{cases}
$$

Dann gilt:

- $T(n) \in O(n)$ falls $d < b$
- $T(n) \in O(n * log(n))$ falls $d = b$
- $T(n) \in O(n^{log_b(d)})$ falls $d > b$

Es ändert sich also nur die Voraussetzung. Hier sind das Abrunden $[n/b]$ und kleinere Abweichungen $n/b + e$ erlaubt.

## Beispiel Merge-Sort
Beim Mergesort Algorithmus wird ein Stapel von Records sortiert. Dabei geht man in drei Schritten vor:<br>
1. Der Stapel wird in zwei (fast) gleich große Teile aufgeteilt; wenn der Stapel nur einem einzigen Record enthält, bricht man ab.
2. Jeder der beiden Hälften wird getrennt sortiert (rekursiv, ebenfalls mit Mergesort).
3. Abschließend werden die beiden sortierten Hälften zu einem sortierten Stapel zuammengemischt.

Sei $T(n)$ die Komplexität, um einen Stapel der Größe $n$ zu sortieren.<br>
Für Schritt 1 und Schritt 3 ist die Komplexität je proportional zu $n$, also zusammen von der Größe $O(n)$.<br>
Für Schritt 2 ist die Komplexität $2*T([\frac{n}{2}])$

Es ergibt sich also

$$T(n) = 2 * T([\frac{n}{2}]) + O(n)$$

Das Mastertheorem wird uns Auskunft darüber geben, wie gross $T(n)$ in diesem Falle ist.

Es gilt also $b = d = 2$ und daraus folgt $T(n) \in O(n)$.
