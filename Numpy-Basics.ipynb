## Numpy Arrays

Um Numpy zu verwenden, muss es importiert werden:

```python
import numpy as np
```

(Numpy als `np` zu importieren ist natürlich nicht zwingend, 
 aber das Kürzel `np` ist fast ein Standard.)

Aus Python-Listen können nun Numpy-Arrays erzeugt werden:

```python
arr = np.array([10, 20, 30])
arr
```

In vielerlei Hinsicht verhält sich ein Numpy-Array wie eine einfache Liste; insbesondere kann die Länge mit `len` angefragt werden, und auf einzelne Elemente kann mt eckigen Klammern zugegriffen werden:

```python
len(arr)
```

```python
arr[0]
```

```python
arr[1]
```

```python
arr[-1]
```

```python
arr[2] = 40
arr
```

## Operatoren für Arrays

Die meisten Methoden für Listen können auch für Arrays verwendet werden. Die Operatoren unterscheiden sich jedoch: Die arithmetischen Operationen `'+'`, `'-'`, `'*'` und`'/'` werden komponentenweise durchgeführt:

```python
np.array([10, 20, 30]) + np.array([100, 200, 300])
```

```python
np.array([10, 20, 30]) - np.array([100, 200, 300])
```

```python
np.array([1, 2, 3]) * np.array([10, 20, 30])
```

```python
np.array([10, 20, 30]) / np.array([1, 2, 4])
```

```python
np.array([10, 20, 30]) // np.array([1, 2, 4])
```

Auf Python-Listen ist von diesen Operationen nur `'+'` definiert, hat aber eine andere Bedeutung, nämlich die Konkatenation:

```python
[10, 20, 30] + [100, 200, 300]
```

Neben den arithmetischen Operationen werden auch die Vergleichsoperationen komponentenweise durchgeführt (im Gegensatz zu den Operationen mit Python-Listen):

```python
np.array([1, 2, 3]) == np.array([3, 2, 1])
```

```python
np.array([1, 2, 3]) != np.array([3, 2, 1])
```

```python
np.array([1, 2, 3]) < np.array([3, 2, 1])
```

```python
np.array([1, 2, 3]) <= np.array([3, 2, 1])
```

Die Operatioren können auch zwischen Skalaren und Arrays verwendet werden:

```python
10 + np.array([1, 2, 3])
```

```python
10 * np.array([1, 2, 3])
```

```python
1 / np.array([1, 2, 4])
```

```python
np.array([1, 2, 4]) / 2
```

```python
np.array([1, 2, 3]) ** 2
```

```python
1 == np.array([1, 2, 3])
```

Auch hier ist das Verhalten anders als bei Python-Listen, z.B. 

```python
3 * [1, 5, 10]
```

Für die meisten Funktionen aus dem `math`-Modul gibt es eine entsprechende Funktion in np, 
die elementweise auf die Elemente eines Arrays angewendet werden kann:

```python
zahlen = np.array([1,2,3,4])
np.sqrt(zahlen)
```

(Ohne Numpy müsste man hier einen `for`-Loop schreiben (oder List-Comprehension verwenden); Numpy ist hier nicht nur einfacher lesbar, sondern bei großen Arrays auch deutlich effizienter.)

## Bereiche

Analog zur Python-range-Funktion gibt es mit `np.arrange` auch eine Range-Funktion für Numpy:

```python
np.arange(5)
```

```python
np.arange(2, 5)
```

Hier gibt es auch eine Version mit einer Schrittweite:

```python
np.arange(0, 10, 2)
```

```python
np.arange(0, 1, 0.1)
```

Eine häufig verwendete Variante ist `np.linspace`: 
hier gibt man nicht die Schrittwite an, sondern die Anzahl Elemente:

```python
np.linspace(0, 1, 6)
```

(Man beachte, dass bei `np.linspace(a, b, n)` die obere Schranke `b` ein Bestandteil des Resultats ist, im Gegensatz zu `range(a, b)`.)

Mit Numpy kann man einfach Wertetabellen von Funktionen erstellen, z.B. für die Funktion 
$y = \sqrt{x} + 3 x^2$ im Bereich $[0,1]$:

```python
xx = np.linspace(0, 1, 11); xx
```

```python
yy = np.sqrt(xx) + 3 * xx ** 2; yy
```

### Ausblick
Diese Wertetabelen werden gebraucht, um Funktion zu Plotten (wird anschließend thematisiert):

```python
import matplotlib.pyplot as plt
xx = np.linspace(0, 1, 501)
yy = np.sqrt(xx) + 3 * xx ** 2
plt.plot(xx, yy)
plt.show()
```
