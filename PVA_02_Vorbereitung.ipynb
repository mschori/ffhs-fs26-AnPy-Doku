# Zahlendarstellungen – Zusammenfassung

## 1. Natürliche Zahlen

### 1.1 Stellenwertsysteme

Das Stellenwertsystem ist die gebräuchlichste Methode, um Zahlen darzustellen. Also Dezimal, Binär, Oktal, Hexadezimal.

Eine Zahl im Stellenwertsystem zur Basis $ b \ge 2 $ hat die Form

$$(a_n a_{n-1} \dots a_0)_b$$

und bedeutet mathematisch:

$$
\sum_{k=0}^{n} a_k \cdot b^k
$$

mit $ 0 \le a_k < b $. (Jede einzelne Ziffer $a_k$ muss kleiner als die Basis $b$ sein. Zum Beispiel: In Dezimal dürfen die Ziffern nur von 0 bis 9 reichen, in Binär nur 0 und 1, in Oktal nur 0 bis 7, und in Hexadezimal von 0 bis 9 sowie A bis F.)

Beispiel (Dezimal):

$$
2345 = 2 \cdot 10^3 + 3 \cdot 10^2 + 4 \cdot 10^1 + 5
$$

Hinweis zur Deutung der Darstellung:

- $b$ = die Basis des Zahlensystems (z.B. 10 für Dezimal, 2 für Binär)
- $a_k$ = die Ziffer an der k-ten Stelle (z.B. 0–9 für Dezimal, 0–1 für Binär)
- $n$ = die Anzahl der Stellen minus 1 (z.B. 3 für eine vierstellige Zahl)
- $k$ = Laufindex für die einzelnen Stellen (beginnend bei 0 für die Einserstelle)
- Jede Stelle hat einen Wert
- - ganz rechts: Einerstelle = $10^0$
- - dann Zehnerstelle = $10^1$
- - dann Hunderterstelle = $10^2$
- - dann Tausenderstelle = $10^3$
- - deshalb $2 * 1000 + 3 * 100 + 4 * 10 + 5 * 1$
- Das Summenzeichen bedeutet hier: "Addiere alle Stellenwerte von rechts nach links."
- Beim Beispiel ist die erste Zahl (Index 0) die $2$ und muss mit $10^3$ multipliziert werden, die zweite Zahl (Index 1) ist die $3$ und muss mit $10^2$ multipliziert werden, usw.

Anderes Beispiel: $(11011)_2$:
- $1 \cdot 2^4 + 1 \cdot 2^3 + 0 \cdot 2^2 + 1 \cdot 2^1 + 1 \cdot 2^0 = 16 + 8 + 0 + 2 + 1 = 27$

Wichtige Basen:

- **Binärsystem (b = 2)**
  Beispiel:
  $$
  1011_2 = 11_{10}
  $$

- **Oktalsystem (b = 8)**
  Ziffern 0–7

- **Hexadezimalsystem (b = 16)**
  Ziffern 0–9 und A–F

Computer rechnen intern binär. Hexadezimal dient als kompakte Darstellung von Bitmustern.

---

### 1.2 Umrechnung von Zahlendarstellungen

Zwei Aufgaben:

1. Darstellung → Zahl
2. Zahl → Darstellung

---

### 1.2.1 Darstellung → Zahl

Direkte Methode:

$$
\sum_{k=0}^{n} a_k \cdot b^k
$$

Horner-Schema (effizienter):

$$
z_{k+1} = z_k \cdot b + a_{k+1}
$$

Beispiel (Basis 8, 2345₈):

- 2
- 2·8 + 3 = 19
- 19·8 + 4 = 156
- 156·8 + 5 = 1253

Rechenaufwand:

- Direkt: n Additionen, 2n−1 Multiplikationen
- Horner: n Additionen, n Multiplikationen

Hinweise zur Deutung der Buchstaben:
- $b$ = Basis des Zahlensystems
- $a_{k+1}$ = Nächste Ziffer der Zahl
- $z_k$ = Zwischenergebnis nach k Schritten (bisher berechneter Zwischenwert)
- $z_{k+1}$ = Aktualisiertes Zwischenergebnis nach Einbeziehung der nächsten Ziffer
- $k$ = Laufindex, der von 0 bis n−1 läuft (für n Ziffern)

---

### 1.2.2 Zahl → Darstellung

Fortgesetzte Division mit Rest.

Beispiel: 1253 in Basis 8

1253 : 8 = 156 Rest 5
156  : 8 = 19  Rest 4
19   : 8 = 2   Rest 3
2    : 8 = 0   Rest 2

Reste von unten nach oben lesen → 2345₈

---

### 1.2.3 Direkte Umrechnung 2 ↔ 8 ↔ 16

- Oktal = 3 Bits
- Hexadezimal = 4 Bits

Binär → Oktal: 3er-Gruppen
Binär → Hex: 4er-Gruppen
Oktal → Binär: jede Ziffer als 3 Bits
Hex → Binär: jede Ziffer als 4 Bits

Hexadezimal ist eine komprimierte Binärdarstellung.

---

## 2. Ganze Zahlen

Darstellung in n Bit.

---

### 2.1 Vorzeichen + Absolutbetrag

Erstes Bit = Vorzeichen.

Darstellbar:

$$
-(2^{n-1}-1) \le z \le 2^{n-1}-1
$$

Probleme:

- +0 und −0 existieren
- Addition benötigt Fallunterscheidung

---

### 2.2 Einerkomplement

Negation = Bitweise Invertierung.<br>
Heisst: Jede 0 wird zu 1, und jede 1 wird zu 0. --> nur beim Einerkomplement!

Weil jede Zahl mit ihrer invertierten Form zusammen immer nur Einsen ergibt:
$$0101 + 1010 = 1111$$

Und 1111 entspricht bei 4 Bit: $2^4 - 1 = 15$.

Somit gilt allgemein: $a + \text{Einerkomplement}(a) = 2^n - 1$.

Dabei bedeutet:
- $a$ = ursprüngliche Zahl
- $n$ = Anzahl der Bits
- Bei 4 Bit also: $2^4 - 1 = 15$

Nachteile:

- Zwei Darstellungen für 0 (0000 und 1111 - bei 4 Bit - beide repräsentieren 0)
- Addition kompliziert, weil 3 (0011) und -3 (1100) gleich 1111 ergibt, was -0 darstellt

---

### 2.3 Zweierkomplement

Standard in Computern.

Vorgehen:

1. Einerkomplement bilden
2. +1 addieren

Darstellbar:

$$
-2^{n-1} \le z \le 2^{n-1}-1
$$

Beispiel in 4 Bit:<br>
5 (0101) → Einerkomplement: 1010 → +1 → 1011 = -5<br>
Weil: Beim Zweierkomplement erhält das linke Bit eine Gewichtung von $-2^{n-1}$, während die anderen Bits positive Gewichtungen haben. In diesem Fall (4 Bit) bedeutet das ($-2^3, 2^2, 2^1, 2^0$):
- Das linke Bit (1) trägt $-2^3 = -8$ bei
- Die anderen Bits (0, 1, 1) tragen $0 \cdot 2^2 + 1 \cdot 2^1 + 1 \cdot 2^0 = 0 + 2 + 1 = 3$ bei
- Insgesamt ergibt das: $-8 + 0 + 2 + 1 = -5$

Eigenschaften:

- Nur eine Darstellung für 0
- Addition funktioniert ohne Vorzeichenunterscheidung
- Subtraktion:
  $$
  a - b = a + (-b)
  $$

Negative Zahl z entspricht:

$$
z + 2^n
$$

Interpretation als Restklassenarithmetik modulo $2^n$.

---

## 3. Brüche

### 3.1 Darstellung

Allgemein:

$$
\sum_{k=-m}^{n} a_k \cdot b^k
$$

Beispiel:

$$
123.95 = 1 \cdot 10^2 + 2 \cdot 10^1 + 3 + 9 \cdot 10^{-1} + 5 \cdot 10^{-2}
$$

---

### Endliche Darstellung

Ein gekürzter Bruch $ \frac{d}{e} $ ist in Basis b genau dann endlich darstellbar, wenn alle Primfaktoren des Nenners (e) auch in der Primfaktorzerlegung von b vorkommen.

Beispiel für $\frac{1}{5}$:

- 1 = d
- 5 = e
- Basis 10 = 2·5 → Nur wenn Nenner durch 2 oder 5 (2er oder 5er Potenzen) teilbar ist, ist die Zahl endlich darstellbar.
- Basis 2, 8, 16 → Nenner muss Zweierpotenz (durch 2 teilbar) sein, damit die Zahl endlich darstellbar ist.

> Merksatz: Ein Bruch endet genau dann, wenn sein Nenner vollständig in die Basis hineinpasst.

Deshalb:

$$
0.2 = \frac{1}{5}
$$

ist im Binärsystem nicht endlich darstellbar.

---

### 3.2 Periodische Brüche

Bei einer Division kann das Ergebnis nur in zwei Verhaltensweisen enden. Entweder es hört irgendwann auf, oder es beginnt sich zu wiederholen. Ein dritter Weg ist nicht möglich. Bei einem Nenner von 22 können die Reste nur 0, 1, 2, ..., 21 sein. Sobald ein Rest zum zweiten Mal auftritt, beginnt die Wiederholung. So entsteht eine Periode.

Division erzeugt:

- Endlichen Bruch oder
- Periodischen Bruch

Satz:

> Eine Zahl ist rational genau dann, wenn ihre Darstellung endlich oder periodisch ist.

Rational bedeutet: Darstellbar als Bruch $ \frac{d}{e} $ mit ganzen Zahlen d und e (e ≠ 0).

Beispiel:

$$
0.13636\ldots = \frac{3}{22}
$$

Beispiel - umgekehrt gilt auch:

- 0.333...
- Setze $x = 0.333...$
- $10x = 3.333...$
- Subtrahiere: $10x - x = 3.333... - 0.333... = 3$
- Somit ist $9x = 3$

Also ist die Periode wieder ein Bruch.

> Irrationale Zahlen habe keine Periode. Zum Beispiel: $\pi = 3.1415926535...$, $e = 2.718281828459045...$, $\sqrt{2} = 1.41421356...$.

---

### 3.3 Bruch → Darstellung

Das ist das Standardverfahren, um einen Bruchteil in ein anderes Zahlensystem umzuwandeln. Beim Ganzzahlanteil teilt man immer durch die Basis; beim Nachkommateil macht man das Gegenteil: man multipliziert immer wieder mit der Basis.

Beispiel $0.2_{10}$ zu Binär:

$$
0.2_{10} = 0.001100110011\ldots_2
$$

Weil:

- 2 * 2 = 0.4 → Ganzzahlanteil: 0, Nachkommateil: 0.4 = 0
- 0.4 * 2 = 0.8 → Ganzzahlanteil: 0, Nachkommateil: 0.8 = 0
- 0.8 * 2 = 1.6 → Ganzzahlanteil: 1, Nachkommateil: 0.6 = 1
- 0.6 * 2 = 1.2 → Ganzzahlanteil: 1, Nachkommateil: 0.2 = 1
- Rest ist wieder 2; daher beginnt hier die Periode von vorne.
- Die Periode ist also 0011
- Darstellung in Binär ist also: $0.001100110011\ldots_2$ (Periode 0011)

Beim Ganzzahlanteil wäre das wie folgt ($13_{10}$ zu Binär):

- 13 : 2 = 6 Rest 1
- 6 : 2 = 3 Rest 0
- 3 : 2 = 1 Rest 1
- 1 : 2 = 0 Rest 1
- Reste von unten nach oben lesen → $1101_2$

## 4. Gleitkommazahlen

### 4.1 Form

$$
v \cdot m \cdot b^e
$$

- v = Vorzeichen (+1 für positive Zahlen, -1 für negative Zahlen - ($v \in {+1, -1}$)
- m = Mantisse (Der eigentliche Zahlenwert, der normalisiert wird, damit er zwischen 1 und b liegt)
- b = Basis (typischerweise 2 für Computer)
- e = Exponent (Besagt, wie stark die Zahl mit der Basis verschoben wird)

Beispiel:

- $3450_{10}
- schreibt man als $3.45 \cdot 10^3$ (hier ist $v=+1$, $m=3.45$, $b=10$, $e=3$)
- - Normalisiert: Komma so verschieben, dass die Mantisse zwischen 1 und b liegt.
- - Komma 3 Stellen nach links verschoben; daher $e=3$.
- Das stimmt weil $3.45 \cdot 10^3 = 3450$

Normalisiert:

$$
1 \le m < b
$$

> Der Computer macht das, weil er so sehr grosse und sehr kleine Zahlen speichern kann. Statt jede Zahl vollständig abzulegen, speichert er Mantisse, Exponent und Vorzeichen. Das spart Speicherplatz. --> Statt 123000000 als Ganzzahl zu speichern, speichert er 1.23 (Mantisse) und 8 (Exponent) und weiß, dass die Basis 10 ist.

> Oder bei sehr kleiner Zahl: 0.0000000045 wird zu 4.5 (Mantisse) und -9 (Exponent) mit Basis 10, also $4.5 \cdot 10^{-7}$.

---

### 4.2 IEEE 754

Nach dieser Norm besteht eine Zahl aus Bits für Vorzeichen, Exponent und Mantisse.

Basis: 2

| Format | Bits | Exponent | Mantisse | Bias |
|--------|------|----------|----------|------|
| float  | 32   | 8        | 23       | 127  |
| double | 64   | 11       | 52       | 1023 |

Exponent gespeichert als:

$$
E = e + \text{Bias}
$$

Mantisse: führende 1 wird nicht gespeichert (normalisierte Zahlen).

> Weil die Matisse endlich lang ist, können unendliche Binärbrüche nicht exakt dargestellt werden. Zum Beispiel: 0.1 im Dezimalsystem ist eine unendliche periodische Zahl im Binärsystem, daher kann sie nicht genau als float oder double gespeichert werden. Daher schneidet der Computer ab und somit entsteht ein Rundungsfehler.

---

### Spezialwerte

Das folgende wird benötigt, um extrem kleine Zahlen (nahe bei 0) und extrem grosse Zahlen (nahe ±∞) darzustellen. Sonst gäbe es Lücken im Zahlenbereich. Das wird auch "gradual underflow" genannt.

- E = 0 → nicht normalisierte Zahlen
- E = Maximalwert → ±∞ oder NaN

Beispiele:

$$
1/(+0) = +\infty
$$

Denn: Division durch Null → Unendlich

$$
0 \cdot \infty = \text{NaN}
$$

Denn: Unbestimmte Form. Man weiss mathematisch nicht eindeutig, was daraus werden soll.

> Weil Infinity noch eine geordnete mathematische Richtung hat, wird zwischen Infinity und NaN unterschieden. NaN ist "Not a Number" und repräsentiert undefinierte oder unbestimmbare Werte, während Infinity eine definierte Richtung (positiv oder negativ) hat, aber dennoch unendlich groß ist.

> Sobald bei einer Berechnung NaN beteiligt ist (z.B. 5 + NaN) wird das Ergebnis ebenfalls NaN, da die Unbestimmtheit weitergegeben wird.

Zu merken:

- $E = 0$ -> sehr kleine Zahlen
- $E = \text{max}, M = 0$ -> ±∞ (Infinity)
- $E = \text{max}, M \neq 0$ -> NaN (Not a Number)

---

### Genauigkeit

float:
- ~7–8 Dezimalstellen
- Bereich ca. $10^{-38}$ bis $10^{38}$

double:
- ~15–16 Dezimalstellen
- Bereich ca. $10^{-308}$ bis $10^{308}$

---

## Praktische Konsequenzen

- 0.2 ist im Binärsystem periodisch → Rundungsfehler
- float/double ungeeignet für Geld
- Besser: int (Rappen) oder BigDecimal
- Zweierkomplement vereinfacht Hardwarearithmetik
- Hexadezimal = kompakte Binärdarstellung

# Konvergenzkriterien für Reihen – Erklärung ohne Formeln

Link zu Moodle: https://moodle.ffhs.ch/mod/page/view.php?id=5151030

## Worum geht es überhaupt?

Eine **Reihe** ist eine unendliche Summe.

Man addiert also unendlich viele Zahlen hintereinander:

a1 + a2 + a3 + a4 + ...

Die zentrale Frage lautet:

> Ergibt diese unendliche Addition einen endlichen Wert – oder wächst sie ins Unendliche?

Wenn sie einen endlichen Wert ergibt, nennt man sie **konvergent**.
Wenn sie unbegrenzt wächst oder keinen festen Wert annimmt, nennt man sie **divergent**.

Die vorgestellten Kriterien helfen dabei, genau das zu entscheiden.

---

## Geometrische Reihe – das wichtigste positive Beispiel

Eine geometrische Reihe entsteht, wenn jeder Summand durch Multiplikation mit einer festen Zahl aus dem vorherigen entsteht.

Beispielidee:
1 + 1/2 + 1/4 + 1/8 + 1/16 + ...

Jeder Schritt wird halbiert.

Sinn dahinter:

Wenn der Faktor betragsmäßig kleiner als 1 ist, werden die Summanden immer kleiner und kleiner – und die Summe stabilisiert sich.
Sie nähert sich einem festen Wert an.

→ Solche Reihen sind konvergent.

Das ist eines der wichtigsten Grundbeispiele.

---

## Harmonische Reihe – das berühmte Gegenbeispiel

Die harmonische Reihe sieht harmlos aus:

1 + 1/2 + 1/3 + 1/4 + 1/5 + ...

Die Summanden werden zwar kleiner, aber sie werden nicht schnell genug kleiner.

Das Entscheidende:

Nur weil die einzelnen Summanden gegen 0 gehen, heißt das noch nicht, dass die ganze Summe konvergiert.

Die harmonische Reihe wächst unbegrenzt weiter.
Sie ist divergent.

Das ist eine wichtige Warnung:
"Summanden gehen gegen 0" reicht nicht als Kriterium.

---

## Leibniz-Kriterium – für alternierende Reihen

Eine **alternierende Reihe** ist eine Reihe, bei der die Vorzeichen wechseln:

+ − + − + − ...

Beispielidee:
1 − 1/2 + 1/3 − 1/4 + 1/5 − ...

Das Leibniz-Kriterium sagt:

Eine solche Reihe konvergiert, wenn zwei Bedingungen erfüllt sind:

1. Die Beträge der Summanden werden immer kleiner.
2. Sie gehen gegen 0.

Warum funktioniert das?

Weil sich die positiven und negativen Beiträge gegenseitig ausgleichen.
Die Summe „schwingt“ sich immer näher an einen festen Wert heran.

Wichtig:
Das ist ein Spezialkriterium nur für Reihen mit Vorzeichenwechsel.

---

## Absolute Konvergenz

Eine Reihe heißt **absolut konvergent**, wenn sie auch dann noch konvergiert, wenn man alle Vorzeichen positiv macht.

Das ist ein sehr starkes Kriterium.

Absolute Konvergenz bedeutet:
Die Reihe ist „robust konvergent“.

---

## Majorantenkriterium – Vergleich nach oben

Hier wird verglichen.

Idee:

Wenn eine Reihe kompliziert aussieht, vergleicht man sie mit einer anderen Reihe, die:

- größer ist
- und von der man weiß, dass sie konvergiert

Wenn selbst die größere Reihe konvergiert, dann muss auch die kleinere konvergieren.

Man nennt die größere Reihe die **Majorante**.

Gedanke dahinter:

Wenn etwas unter einer konvergenten „Decke“ bleibt, kann es selbst nicht ins Unendliche wachsen.

---

## Minorantenkriterium – Vergleich nach unten

Das ist das Gegenstück.

Wenn man eine Reihe findet, die:

- kleiner ist
- aber trotzdem divergiert

dann muss auch die größere divergieren.

Die kleinere divergente Reihe nennt man **Minorante**.

Gedanke:

Wenn selbst etwas Kleineres schon ins Unendliche wächst, dann wächst das Größere erst recht.

---

## Quotientenkriterium – Wachstumsvergleich

Hier schaut man sich an, wie schnell die Summanden wachsen oder schrumpfen.

Man vergleicht jeweils zwei aufeinanderfolgende Summanden.

Wenn der nächste Summand dauerhaft deutlich kleiner wird als der vorherige, dann schrumpfen die Beiträge schnell genug.

Dann konvergiert die Reihe.

Das Kriterium ist besonders nützlich bei Reihen mit:

- Potenzen $\to k^2 \to$ variable steht unten als Basis
- Fakultäten $\to k!$
- Exponentialfunktionen $\to 2^k \to$ variable steht oben im Exponenten

Gedanke dahinter:

Es geht nicht nur darum, ob etwas kleiner wird, sondern **wie schnell** es kleiner wird.

---

## Der große Zusammenhang

Man kann sich alle Kriterien so merken:

Konvergenz hängt davon ab, wie schnell die Summanden gegen 0 gehen.

- Gehen sie sehr schnell gegen 0 → meist konvergent.
- Gehen sie langsam gegen 0 → oft divergent.
- Wechseln die Vorzeichen → Ausgleich möglich.
- Vergleich mit bekannten Reihen hilft enorm.

---

## Typische Prüfungsstrategie

1. Zuerst prüfen: Gehen die Summanden überhaupt gegen 0?
2. Erkennen: Ist es eine geometrische Reihe?
3. Wechseln die Vorzeichen? → Leibniz prüfen.
4. Vergleich mit bekannter Reihe? → Majorante oder Minorante.
5. Exponential-/Fakultätsstruktur? → Quotientenkriterium.

---

## Intuitive Kernaussage

Eine unendliche Summe ist wie ein ewiges Nachgießen in ein Glas.

- Tropft es immer langsamer → Glas füllt sich bis zu einem festen Niveau.
- Tropft es zu stark weiter → Glas läuft über.

Die Konvergenzkriterien sind Werkzeuge, um vorherzusagen, was passiert.
