# Zahlendarstellungen – Zusammenfassung

## 1. Natürliche Zahlen

### 1.1 Stellenwertsysteme

Eine Zahl im Stellenwertsystem zur Basis $ b \ge 2 $ hat die Form

$$(a_n a_{n-1} \dots a_0)_b$$

und bedeutet mathematisch:

$$
\sum_{k=0}^{n} a_k \cdot b^k
$$

mit $ 0 \le a_k < b $.

Beispiel (Dezimal):

$$
2345 = 2 \cdot 10^3 + 3 \cdot 10^2 + 4 \cdot 10^1 + 5
$$

Wichtige Basen:

- **Binärsystem (b = 2)**
  Beispiel:
  $$
  1011_2 = 11_{10}
  $$

- **Oktalsystem (b = 8)**
  Ziffern 0–7

- **Hexadezimalsystem (b = 16)**
  Ziffern 0–9 und A–F

Computer rechnen intern binär. Hexadezimal dient als kompakte Darstellung von Bitmustern.

---

### 1.2 Umrechnung von Zahlendarstellungen

Zwei Aufgaben:

1. Darstellung → Zahl
2. Zahl → Darstellung

---

### 1.2.1 Darstellung → Zahl

Direkte Methode:

$$
\sum_{k=0}^{n} a_k \cdot b^k
$$

Horner-Schema (effizienter):

$$
z_{k+1} = z_k \cdot b + a_{k+1}
$$

Beispiel (Basis 8, 2345₈):

- 2
- 2·8 + 3 = 19
- 19·8 + 4 = 156
- 156·8 + 5 = 1253

Rechenaufwand:

- Direkt: n Additionen, 2n−1 Multiplikationen
- Horner: n Additionen, n Multiplikationen

---

### 1.2.2 Zahl → Darstellung

Fortgesetzte Division mit Rest.

Beispiel: 1253 in Basis 8

1253 : 8 = 156 Rest 5
156  : 8 = 19  Rest 4
19   : 8 = 2   Rest 3
2    : 8 = 0   Rest 2

Reste von unten nach oben lesen → 2345₈

---

### 1.2.3 Direkte Umrechnung 2 ↔ 8 ↔ 16

- Oktal = 3 Bits
- Hexadezimal = 4 Bits

Binär → Oktal: 3er-Gruppen
Binär → Hex: 4er-Gruppen
Oktal → Binär: jede Ziffer als 3 Bits
Hex → Binär: jede Ziffer als 4 Bits

Hexadezimal ist eine komprimierte Binärdarstellung.

---

## 2. Ganze Zahlen

Darstellung in n Bit.

---

### 2.1 Vorzeichen + Absolutbetrag

Erstes Bit = Vorzeichen.

Darstellbar:

$$
-(2^{n-1}-1) \le z \le 2^{n-1}-1
$$

Probleme:

- +0 und −0 existieren
- Addition benötigt Fallunterscheidung

---

### 2.2 Einerkomplement

Negation = Bitweise Invertierung.

Eigenschaft:

$$
a + \text{Einerkomplement}(a) = 2^n - 1
$$

Nachteile:

- Zwei Darstellungen für 0
- Addition kompliziert

---

### 2.3 Zweierkomplement

Standard in Computern.

Vorgehen:

1. Einerkomplement bilden
2. +1 addieren

Darstellbar:

$$
-2^{n-1} \le z \le 2^{n-1}-1
$$

Eigenschaften:

- Nur eine Darstellung für 0
- Addition funktioniert ohne Vorzeichenunterscheidung
- Subtraktion:
  $$
  a - b = a + (-b)
  $$

Negative Zahl z entspricht:

$$
z + 2^n
$$

Interpretation als Restklassenarithmetik modulo $2^n$.

---

## 3. Brüche

### 3.1 Darstellung

Allgemein:

$$
\sum_{k=-m}^{n} a_k \cdot b^k
$$

Beispiel:

$$
123.95 = 1 \cdot 10^2 + 2 \cdot 10^1 + 3 + 9 \cdot 10^{-1} + 5 \cdot 10^{-2}
$$

---

### Endliche Darstellung

Ein gekürzter Bruch $ \frac{d}{e} $ ist in Basis b genau dann endlich darstellbar, wenn die Primfaktoren von e nur in der Primfaktorzerlegung von b vorkommen.

Beispiele:

- Basis 10 = 2·5 → Nenner nur 2 und 5
- Basis 2, 8, 16 → Nenner muss Zweierpotenz sein

Deshalb:

$$
0.2 = \frac{1}{5}
$$

ist im Binärsystem nicht endlich darstellbar.

---

### 3.2 Periodische Brüche

Division erzeugt:

- Endlichen Bruch oder
- Periodischen Bruch

Satz:

> Eine Zahl ist rational genau dann, wenn ihre Darstellung endlich oder periodisch ist.

Beispiel:

$$
0.13636\ldots = \frac{3}{22}
$$

---

### 3.3 Bruch → Darstellung

Vorgehen:

1. Ganzzahlanteil bestimmen
2. Bruchanteil wiederholt mit b multiplizieren
3. Ganzanteil jedes Produkts = nächste Stelle

Beispiel:

$$
0.2_{10} = 0.001100110011\ldots_2
$$

Periode = 0011

---

## 4. Gleitkommazahlen

### 4.1 Form

$$
v \cdot m \cdot b^e
$$

- v = Vorzeichen
- m = Mantisse
- e = Exponent

Normalisiert:

$$
1 \le m < b
$$

---

### 4.2 IEEE 754

Basis: 2

| Format | Bits | Exponent | Mantisse | Bias |
|--------|------|----------|----------|------|
| float  | 32   | 8        | 23       | 127  |
| double | 64   | 11       | 52       | 1023 |

Exponent gespeichert als:

$$
E = e + \text{Bias}
$$

Mantisse: führende 1 wird nicht gespeichert (normalisierte Zahlen).

---

### Spezialwerte

- E = 0 → nicht normalisierte Zahlen
- E = Maximalwert → ±∞ oder NaN

Beispiele:

$$
1/(+0) = +\infty
$$

$$
0 \cdot \infty = \text{NaN}
$$

---

### Genauigkeit

float:
- ~7–8 Dezimalstellen
- Bereich ca. $10^{-38}$ bis $10^{38}$

double:
- ~15–16 Dezimalstellen
- Bereich ca. $10^{-308}$ bis $10^{308}$

---

## Praktische Konsequenzen

- 0.2 ist im Binärsystem periodisch → Rundungsfehler
- float/double ungeeignet für Geld
- Besser: int (Rappen) oder BigDecimal
- Zweierkomplement vereinfacht Hardwarearithmetik
- Hexadezimal = kompakte Binärdarstellung

# Konvergenzkriterien für Reihen – Erklärung ohne Formeln

Link zu Moodle: https://moodle.ffhs.ch/mod/page/view.php?id=5151030

## Worum geht es überhaupt?

Eine **Reihe** ist eine unendliche Summe.

Man addiert also unendlich viele Zahlen hintereinander:

a1 + a2 + a3 + a4 + ...

Die zentrale Frage lautet:

> Ergibt diese unendliche Addition einen endlichen Wert – oder wächst sie ins Unendliche?

Wenn sie einen endlichen Wert ergibt, nennt man sie **konvergent**.
Wenn sie unbegrenzt wächst oder keinen festen Wert annimmt, nennt man sie **divergent**.

Die vorgestellten Kriterien helfen dabei, genau das zu entscheiden.

---

## Geometrische Reihe – das wichtigste positive Beispiel

Eine geometrische Reihe entsteht, wenn jeder Summand durch Multiplikation mit einer festen Zahl aus dem vorherigen entsteht.

Beispielidee:
1 + 1/2 + 1/4 + 1/8 + 1/16 + ...

Jeder Schritt wird halbiert.

Sinn dahinter:

Wenn der Faktor betragsmäßig kleiner als 1 ist, werden die Summanden immer kleiner und kleiner – und die Summe stabilisiert sich.
Sie nähert sich einem festen Wert an.

→ Solche Reihen sind konvergent.

Das ist eines der wichtigsten Grundbeispiele.

---

## Harmonische Reihe – das berühmte Gegenbeispiel

Die harmonische Reihe sieht harmlos aus:

1 + 1/2 + 1/3 + 1/4 + 1/5 + ...

Die Summanden werden zwar kleiner, aber sie werden nicht schnell genug kleiner.

Das Entscheidende:

Nur weil die einzelnen Summanden gegen 0 gehen, heißt das noch nicht, dass die ganze Summe konvergiert.

Die harmonische Reihe wächst unbegrenzt weiter.
Sie ist divergent.

Das ist eine wichtige Warnung:
"Summanden gehen gegen 0" reicht nicht als Kriterium.

---

## Leibniz-Kriterium – für alternierende Reihen

Eine **alternierende Reihe** ist eine Reihe, bei der die Vorzeichen wechseln:

+ − + − + − ...

Beispielidee:
1 − 1/2 + 1/3 − 1/4 + 1/5 − ...

Das Leibniz-Kriterium sagt:

Eine solche Reihe konvergiert, wenn zwei Bedingungen erfüllt sind:

1. Die Beträge der Summanden werden immer kleiner.
2. Sie gehen gegen 0.

Warum funktioniert das?

Weil sich die positiven und negativen Beiträge gegenseitig ausgleichen.
Die Summe „schwingt“ sich immer näher an einen festen Wert heran.

Wichtig:
Das ist ein Spezialkriterium nur für Reihen mit Vorzeichenwechsel.

---

## Absolute Konvergenz

Eine Reihe heißt **absolut konvergent**, wenn sie auch dann noch konvergiert, wenn man alle Vorzeichen positiv macht.

Das ist ein sehr starkes Kriterium.

Absolute Konvergenz bedeutet:
Die Reihe ist „robust konvergent“.

---

## Majorantenkriterium – Vergleich nach oben

Hier wird verglichen.

Idee:

Wenn eine Reihe kompliziert aussieht, vergleicht man sie mit einer anderen Reihe, die:

- größer ist
- und von der man weiß, dass sie konvergiert

Wenn selbst die größere Reihe konvergiert, dann muss auch die kleinere konvergieren.

Man nennt die größere Reihe die **Majorante**.

Gedanke dahinter:

Wenn etwas unter einer konvergenten „Decke“ bleibt, kann es selbst nicht ins Unendliche wachsen.

---

## Minorantenkriterium – Vergleich nach unten

Das ist das Gegenstück.

Wenn man eine Reihe findet, die:

- kleiner ist
- aber trotzdem divergiert

dann muss auch die größere divergieren.

Die kleinere divergente Reihe nennt man **Minorante**.

Gedanke:

Wenn selbst etwas Kleineres schon ins Unendliche wächst, dann wächst das Größere erst recht.

---

## Quotientenkriterium – Wachstumsvergleich

Hier schaut man sich an, wie schnell die Summanden wachsen oder schrumpfen.

Man vergleicht jeweils zwei aufeinanderfolgende Summanden.

Wenn der nächste Summand dauerhaft deutlich kleiner wird als der vorherige, dann schrumpfen die Beiträge schnell genug.

Dann konvergiert die Reihe.

Das Kriterium ist besonders nützlich bei Reihen mit:

- Potenzen $\to k^2 \to$ variable steht unten als Basis
- Fakultäten $\to k!$
- Exponentialfunktionen $\to 2^k \to$ variable steht oben im Exponenten

Gedanke dahinter:

Es geht nicht nur darum, ob etwas kleiner wird, sondern **wie schnell** es kleiner wird.

---

## Der große Zusammenhang

Man kann sich alle Kriterien so merken:

Konvergenz hängt davon ab, wie schnell die Summanden gegen 0 gehen.

- Gehen sie sehr schnell gegen 0 → meist konvergent.
- Gehen sie langsam gegen 0 → oft divergent.
- Wechseln die Vorzeichen → Ausgleich möglich.
- Vergleich mit bekannten Reihen hilft enorm.

---

## Typische Prüfungsstrategie

1. Zuerst prüfen: Gehen die Summanden überhaupt gegen 0?
2. Erkennen: Ist es eine geometrische Reihe?
3. Wechseln die Vorzeichen? → Leibniz prüfen.
4. Vergleich mit bekannter Reihe? → Majorante oder Minorante.
5. Exponential-/Fakultätsstruktur? → Quotientenkriterium.

---

## Intuitive Kernaussage

Eine unendliche Summe ist wie ein ewiges Nachgießen in ein Glas.

- Tropft es immer langsamer → Glas füllt sich bis zu einem festen Niveau.
- Tropft es zu stark weiter → Glas läuft über.

Die Konvergenzkriterien sind Werkzeuge, um vorherzusagen, was passiert.
