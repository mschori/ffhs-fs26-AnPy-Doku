# Aufgaben für Einsteiger aus Dalwigk

5.2 Gib die Summe $0 + 5 + 10 + ... + n * 5$ mit Hilfe des Summenzeichens an und
beweise, dass man ihren Wert für alle n $\in$ N durch 2,5 * n * (n + 1) berechnen kann.

Lösung für Summenzeichen:
$$\sum_{k=0}^{n} 5 * k$$
