## Einleitung

Die Visualisierung von Daten spielt eine zentrale Rolle in der Analyse und Interpretation von Informationen. Visuelle Darstellungen sind ein wesentliches Mittel zur Übermittlung von Informationen, zur Erhellung von Sachverhalten sowie zur Erkennung von Mustern und zur Gewinnung von Einsichten. Insbesondere in praxisnahen Anwendungen ermöglicht eine geeignete grafische Darstellung, komplexe Zusammenhänge verständlich zu kommunizieren und fundierte Entscheidungen zu treffen [1].

Datenvisualisierung bezeichnet dabei die Darstellung von Daten in einem grafischen oder bildlichen Format, wodurch Analyseergebnisse übersichtlich präsentiert und auch komplexe Konzepte besser verstanden werden können. Moderne Technologien ermöglichen darüber hinaus interaktive Visualisierungen, bei denen Nutzende gezielt Daten auswählen und Darstellungen anpassen können, um tiefere Einblicke zu gewinnen [1].

Durch die zunehmende Verfügbarkeit grosser Datenmengen, beispielsweise durch internationale Organisationen, staatliche Institutionen oder Nichtregierungsorganisationen, gewinnt die Fähigkeit zur effektiven Visualisierung zusätzlich an Bedeutung. Während umfangreiche Datenbestände heute leicht zugänglich sind, stellt deren verständliche Aufbereitung eine zentrale Herausforderung dar. Geeignete Visualisierungen leisten hierbei einen entscheidenden Beitrag zum Verständnis komplexer, oft vielschichtiger (multidimensionaler) Zusammenhänge [1] [2].

Ziel dieser Arbeit ist es, eine strukturierte Beispielsammlung verschiedener Diagrammtypen mit der Python-Bibliothek Matplotlib zu erstellen und deren Einsatzgebiete systematisch zu untersuchen. Dabei wird insbesondere analysiert, welcher Diagrammtyp für welche Art von Daten geeignet ist und welche Limitationen bei der Verwendung auftreten können.

Als praxisnahes Szenario wird eine Kombination aus analytischen Funktionen sowie fiktiven Datensätzen verwendet. Die Funktionsgraphen dienen zur Darstellung mathematischer Zusammenhänge, während künstlich erzeugte Datensätze typische Anwendungsfälle aus der Praxis repräsentieren.

Ein besonderer Fokus liegt auf der Interpretation der erzeugten Graphiken sowie der kritischen Diskussion ihrer Aussagekraft. Dadurch soll nicht nur die technische Umsetzung, sondern auch die korrekte Anwendung von Visualisierungsmethoden nachvollziehbar gemacht werden.

## Theoretische Grundlagen

Die Wahl des geeigneten Diagrammtyps ist entscheidend für die korrekte Interpretation von Daten. Unterschiedliche Diagrammarten eignen sich für unterschiedliche Datenstrukturen und Fragestellungen [3][5]. Funktionsgraphen werden verwendet, um analytische Zusammenhänge zwischen Variablen darzustellen. Mehrere Funktionsgraphen in einer Darstellung ermöglichen den direkten Vergleich verschiedener Zusammenhänge. Dies ist insbesondere hilfreich, um Unterschiede im Wachstum oder Verhalten von Funktionen zu erkennen. Balkendiagramme eignen sich zur Darstellung diskreter Kategorien und ermöglichen einen klaren Vergleich einzelner Werte. Tortendiagramme visualisieren Anteile eines Ganzen, sind jedoch nur bei wenigen Kategorien sinnvoll einsetzbar. Histogramme dienen der Darstellung von Verteilungen und werden insbesondere in der statistischen Analyse eingesetzt [4][5].

Ein zentraler Aspekt der Datenvisualisierung ist die korrekte Zuordnung von Diagrammtyp und Datenstruktur. Eine ungeeignete Wahl des Diagrammtyps kann zu Fehlinterpretationen führen und die Aussagekraft der Daten erheblich verzerren [3]. Beispielsweise sind Tortendiagramme bei vielen Kategorien schwer lesbar, während Histogramme für kategoriale Daten ungeeignet sind.

## Programmcode
### Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### CSV-Datensatz erzeugen

```python
def create_sample_data_product_sales() -> pd.DataFrame:
    """
    Creates dataframe for product sales.
    This sample Data is used for the visualization of the Bar-Chart and the Pie-Chart
    :return: pd.DataFrame
    """
    data = {
        "Produkt": ["A", "B", "C", "D", "E"],
        "Verkäufe": [120, 90, 150, 70, 110],
        "Kategorie": ["Elektronik", "Elektronik", "Haushalt", "Haushalt", "Sport"]
    }
    return pd.DataFrame(data)


def create_sample_data_sales_distribution(n: int = 100) -> np.ndarray:
    """
    Creates sales distribution for n years.
    This sample Data is used for the visualization of the histogram.
    :param n: Amount of years
    :return: np.ndarray
    """
    np.random.seed(42)
    return np.random.normal(loc=100, scale=20, size=n)
```

### Funktionsgraphen erstellen

```python
def plot_function() -> None:
    """
    Displays the graph of the function y = x^2.
    """
    x = np.linspace(0, 10, 100)
    y = x ** 2

    plt.figure()
    plt.plot(x, y)
    plt.title("Funktionsgraph: y = x^2")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.show()
```

### Mehrere Funktionsgraphen erstellen

```python
def plot_multiple_functions() -> None:
    """
    Displays the graph of the functions y = x, y = x^2 and y = log(x+1).
    """
    x = np.linspace(0, 10, 100)

    plt.figure()
    plt.plot(x, x, label="y = x")
    plt.plot(x, x ** 2, label="y = x^2")
    plt.plot(x, np.log(x + 1), label="y = log(x+1)")

    plt.title("Vergleich mehrerer Funktionen")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()
```

### Balkendiagramm erstellen

```python
def plot_bar_chart(df: pd.DataFrame) -> None:
    """
    Displays the graph of the bar-chart of the given dataframe.
    :param df: Pandas dataframe to display
    """
    plt.figure()
    plt.bar(df["Produkt"], df["Verkäufe"])

    plt.title("Verkäufe pro Produkt")
    plt.xlabel("Produkt")
    plt.ylabel("Verkäufe")
    plt.show()
```

### Tortendiagramm erstellen

```python
def plot_pie_chart(df: pd.DataFrame) -> None:
    """
    Displays the graph of the pie chart of the given dataframe.
    :param df: Pandas dataframe to display
    """
    category_sum = df.groupby("Kategorie")["Verkäufe"].sum()

    plt.figure()
    plt.pie(category_sum, labels=category_sum.index, autopct="%1.1f%%")

    plt.title("Anteile der Kategorien")
    plt.show()
```

### Histogramm erstellen

```python
def plot_histogram(data: np.ndarray) -> None:
    """
    Displays the histogram of the given data.
    :param data: Numpy array to display
    """
    plt.figure()
    plt.hist(data, bins=10)

    plt.title("Verteilung der täglichen Verkäufe")
    plt.xlabel("Verkäufe")
    plt.ylabel("Häufigkeit")
    plt.grid(True)
    plt.show()
```

Für die Visualisierung der weiteren Daten werden zwei Datensätze erstellt. Das Tortendiagramm zeigt die prozentualen Anteile der einzelnen Kategorien. Die Kategorie Haushalt stellt den grössten Anteil dar, während Sport den kleinsten Anteil ausmacht.

### Histogramm
Das Histogramm zeigt die Verteilung der täglichen Verkaufszahlen. Die x-Achse repräsentiert die Anzahl der Verkäufe pro Tag, während die y-Achse angibt, wie häufig bestimmte Wertebereiche auftreten. ## Diskussion der Ergebnisse

Die Ergebnisse zeigen deutlich, dass die Wahl des Diagrammtyps einen erheblichen Einfluss auf die Interpretierbarkeit der Daten hat, was auch in der Literatur zur Datenvisualisierung betont wird [3], [4].

Funktionsgraphen eignen sich hervorragend zur Darstellung mathematischer Zusammenhänge. Insbesondere der Vergleich mehrerer Funktionen in einer Grafik ermöglicht es, unterschiedliche Wachstumsverhalten direkt zu erkennen. Ohne diese gemeinsame Darstellung wäre ein solcher Vergleich deutlich schwieriger.

Das Balkendiagramm erweist sich als besonders geeignet für den Vergleich diskreter Werte. Die klare Trennung der Kategorien ermöglicht eine schnelle und intuitive Interpretation. Im Gegensatz dazu wäre ein Liniendiagramm für diese Daten weniger geeignet, da es einen kontinuierlichen Zusammenhang suggerieren würde, der nicht vorhanden ist.

Das Tortendiagramm bietet eine anschauliche Darstellung relativer Anteile. Dies bestätigt bekannte Limitationen von Tortendiagrammen, die insbesondere bei vielen Kategorien oder geringen Unterschieden schwer interpretierbar sind [4]. In solchen Fällen wäre ein Balkendiagramm die bessere Wahl.

Das Histogramm ermöglicht eine Analyse der Verteilung der Daten. Es zeigt, wie häufig bestimmte Wertebereiche auftreten, und ist besonders nützlich zur Identifikation von Häufungen oder Ausreissern. Ein Histogramm mit einer sehr kleinen Anzahl von Datenpunkten ist wenig aussagekräftig, da keine sinnvolle Verteilung erkennbar ist. Aus diesem Grund wurden für die Analyse künstliche Daten mit ausreichender Stichprobengrösse erzeugt. Eine ausreichende Stichprobengrösse ist entscheidend, um statistische Verteilungen sinnvoll darzustellen [2].

Die unterschiedlichen Diagrammtypen erfordern unterschiedliche Datenstrukturen. Während Histogramme auf kontinuierlichen Einzelwerten basieren, benötigen Balken- und Tortendiagramme aggregierte kategoriale Daten. Eine direkte Wiederverwendung derselben Daten ist daher nicht sinnvoll. Dies unterstreicht die Notwendigkeit, den Diagrammtyp stets an die zugrunde liegende Datenstruktur anzupassen, um Fehlinterpretationen zu vermeiden [3].

Ein weiterer wichtiger Aspekt ist die Gestaltung der Diagramme. Beschriftungen, Titel und Skalierungen haben einen direkten Einfluss auf die Verständlichkeit. Fehlende oder unklare Beschriftungen können die Aussagekraft erheblich reduzieren.

Limitationen dieser Arbeit bestehen insbesondere in der Verwendung eines kleinen, künstlichen Datensatzes. In realen Anwendungen sind Datensätze oft deutlich grösser und komplexer, was zusätzliche Herausforderungen mit sich bringt, beispielsweise hinsichtlich Performance oder Visualisierungsstrategien.

Ein weiterer zentraler Aspekt ist, dass Visualisierung nicht nur der Darstellung dient, sondern auch die Wahrnehmung der Daten aktiv beeinflusst. Unterschiedliche Diagrammtypen können dieselben Daten unterschiedlich interpretierbar machen, was die Verantwortung bei der Wahl der Darstellung zusätzlich erhöht [3], [4].

Diese Ergebnisse zeigen, dass die korrekte Wahl des Diagrammtyps nicht nur eine technische Entscheidung ist, sondern eine wesentliche Voraussetzung für eine valide Dateninterpretation darstellt.
