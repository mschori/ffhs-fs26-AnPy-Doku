## Einleitung

Die Approximation von Funktionen ist ein zentraler Bestandteil der numerischen Mathematik und spielt in vielen technischen und wissenschaftlichen Anwendungen eine wichtige Rolle. In der Praxis sind exakte Funktionsverläufe häufig nicht direkt bekannt, sondern nur durch eine endliche Anzahl von Mess- oder Stützpunkten gegeben. Um dennoch Aussagen über den Verlauf zwischen diesen Punkten machen zu können, werden Approximationsverfahren eingesetzt.

Eine besonders wichtige Methode zur Beschreibung glatter Kurven zwischen gegebenen Punkten sind Splines. Dabei handelt es sich um stückweise definierte Funktionen, die so miteinander verbunden werden, dass ein gleichmässiger und glatter Kurvenverlauf entsteht. Kubische Splines sind in diesem Zusammenhang besonders relevant, da sie eine gute Balance zwischen mathematischer Einfachheit, Glattheit und Approximationsqualität bieten.

Ziel dieser Arbeit ist es, eine nichtlineare Kurve anhand ausgewählter Stützstellen mit kubischen Splines zu approximieren. Als Beispiel wird die Funktion

$$f(x)=sin(x)+0.3x$$

verwendet. Diese Funktion eignet sich gut für die Untersuchung, da sie einerseits einen trigonometrischen Anteil besitzt und damit die Vorgaben der Aufgabenstellung erfüllt, andererseits durch den linearen Zusatz einen ansteigenden Trend aufweist. Dadurch lassen sich Unterschiede zwischen exakter Funktion und Approximation besser erkennen als bei einer reinen Sinusfunktion.

Im Zentrum der Arbeit stehen die grafische Darstellung der exakten Funktion, der Spline-Approximation und der verwendeten Stützpunkte sowie die Diskussion der Approximationsqualität. Insbesondere wird untersucht, wie sich die Anzahl der Stützstellen auf die Genauigkeit der Approximation auswirkt und welche Möglichkeiten zur Verbesserung der Approximationsgüte bestehen.

## Theoretische Grundlagen

Splines sind stückweise definierte Funktionen, die zur Interpolation oder Approximation gegebener Datenpunkte verwendet werden. Anstatt eine einzige Funktion über das gesamte Intervall zu bestimmen, wird das Intervall in mehrere Teilintervalle zerlegt. Für jedes dieser Teilintervalle wird ein eigenes Polynom definiert. Diese Teilfunktionen werden anschliessend so miteinander verbunden, dass ein zusammenhängender und möglichst glatter Gesamtverlauf entsteht.

In dieser Arbeit werden kubische Splines verwendet. Dabei handelt es sich um Polynome dritten Grades, die zwischen jeweils zwei benachbarten Stützstellen definiert sind. Kubische Splines werden so konstruiert, dass sie an den Verbindungsstellen nicht nur denselben Funktionswert besitzen, sondern auch in der ersten und zweiten Ableitung übereinstimmen. Dadurch entstehen keine Sprünge oder Knicke, sondern ein glatter Kurvenverlauf.

Die Forderung nach Stetigkeit der ersten Ableitung bedeutet, dass die Tangentenrichtung an den Übergangsstellen übereinstimmt. Die zusätzliche Stetigkeit der zweiten Ableitung sorgt dafür, dass auch die Krümmung der Kurve keine abrupten Änderungen aufweist. Diese Eigenschaften machen kubische Splines besonders geeignet für Anwendungen, in denen natürliche oder physikalisch plausible Verläufe beschrieben werden sollen.

Ein wesentlicher Vorteil von Splines gegenüber globalen Polynomen besteht darin, dass sie lokal definiert sind. Wird ein zusätzliches Stützpunktpaar verändert, beeinflusst dies nur einen begrenzten Bereich der Kurve. Bei globalen Polynomen kann dagegen bereits eine kleine Änderung einzelner Punkte den Verlauf der gesamten Funktion stark verändern. Ausserdem neigen Polynome höheren Grades häufig zu unerwünschten Schwingungen zwischen den Stützstellen, während Splines in der Regel einen stabileren Verlauf liefern.

Neben der Wahl der Stützstellen sind auch die Randbedingungen von Bedeutung. Da an den äusseren Intervallgrenzen zusätzliche Bedingungen benötigt werden, um den Spline eindeutig zu bestimmen, werden in dieser Arbeit sogenannte natürliche Randbedingungen verwendet. Dabei wird angenommen, dass die zweite Ableitung an beiden Randpunkten gleich null ist. Anschaulich bedeutet dies, dass die Kurve an den Rändern keine zusätzliche Krümmung erzwungen bekommt. Diese Wahl führt häufig zu einem glatten und plausiblen Randverhalten und ist deshalb für eine verständliche Einführung in kubische Splines gut geeignet.

Die Qualität einer Spline-Approximation hängt wesentlich von der Anzahl und Verteilung der Stützstellen ab. Mit wenigen Stützstellen kann nur ein grober Verlauf der Funktion erfasst werden. Mit zunehmender Anzahl an Stützstellen wird die Approximation in der Regel genauer, da mehr Informationen über den exakten Funktionsverlauf zur Verfügung stehen. Besonders in Bereichen mit stärkerer Krümmung ist eine ausreichend dichte Wahl der Stützstellen wichtig.

Zur Beurteilung der Qualität einer Approximation können unterschiedliche Ansätze verwendet werden. Einerseits ist eine visuelle Beurteilung durch den direkten Vergleich von exakter Funktion und Approximationskurve möglich. Andererseits kann der Fehler zwischen exakter Funktion und Spline an vielen Zwischenpunkten berechnet werden. In dieser Arbeit wird neben der grafischen Darstellung auch eine einfache Fehleranalyse durchgeführt, um die Genauigkeit der Approximation nachvollziehbar zu machen.

Durch die Wahl der Funktion $f(x)=sin(x)+0.3x$ ergibt sich ein anschauliches Beispiel für die Untersuchung. Die Funktion ist glatt und stetig, weist jedoch im Gegensatz zur reinen Sinusfunktion zusätzlich einen linearen Trend auf. Dadurch wird die Approximation etwas anspruchsvoller und Unterschiede zwischen Varianten mit unterschiedlich vielen Stützstellen werden deutlicher sichtbar.

## Implementierungsidee

Die Implementierung orientiert sich an der Struktur der vorherigen Arbeit und folgt einem modularen Ansatz.

Die Umsetzung basiert auf folgenden Komponenten:

- Definition der exakten Funktion (Sinusfunktion)
- Erzeugung von Stützstellen mit unterschiedlicher Anzahl
- Berechnung der Spline-Approximation mit CubicSpline
- Visualisierung der Ergebnisse mit Matplotlib
- Einfache Fehleranalyse durch Vergleich von exakter Funktion und Approximation

Zur Untersuchung des Einflusses der Stützstellen werden drei Varianten betrachtet:

- 4 Stützstellen (grobe Approximation)
- 8 Stützstellen (mittlere Approximation)
- 12 Stützstellen (feinere Approximation)

Für jede Variante wird eine separate Grafik erstellt, die folgende Elemente enthält:

- exakte Funktion
- Spline-Approximation
- Stützpunkte

## Programmcode
### Imports

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import CubicSpline
```

## Definition der exakten Funktion

```python
def exact_function(x: np.ndarray) -> np.ndarray:
    """
    Exact function used for comparison.
    """
    return np.sin(x) + 0.3 * x
```

### Erstellung der Stützstellen

```python
def create_support_points(n: int, start: float, end: float):
    """
    Creates n support points in a given interval.
    """
    x = np.linspace(start, end, n)
    y = exact_function(x)
    return x, y
```

### Erstellung der Splines

```python
def create_spline(x: np.ndarray, y: np.ndarray) -> CubicSpline:
    """
    Creates a cubic spline with natural boundary conditions.
    """
    return CubicSpline(x, y, bc_type='natural')
```

### Plot-Funktion

```python
def plot_spline_approximation(n_points: int, start: float, end: float) -> None:
    """
    Plots the exact function, the spline approximation and the support points.
    """
    x_support, y_support = create_support_points(n_points, start, end)
    spline = create_spline(x_support, y_support)

    x_fine = np.linspace(start, end, 500)
    y_exact = exact_function(x_fine)
    y_spline = spline(x_fine)

    error = np.abs(y_exact - y_spline)
    max_error = np.max(error)

    plt.figure()
    plt.plot(x_fine, y_exact, label="Exakte Funktion sin(x) + 0.3x")
    plt.plot(x_fine, y_spline, linestyle="--", label="Spline Approximation")
    plt.scatter(x_support, y_support, color="red", label="Stützstellen")

    plt.title(f"Spline Approximation mit {n_points} Stützstellen (max. Fehler: {max_error:.4f})")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()
```

## Darstellung der Ergebnisse
Zur Analyse der Approximation wird die Sinusfunktion im Intervall [0,2π] betrachtet.

### Variante 1: 4 Stützstellen
Die Approximation mit 4 Stützstellen zeigt deutliche Abweichungen vom exakten Funktionsverlauf. Aufgrund der grossen Abstände zwischen den einzelnen Punkten kann der Spline die Krümmung der Funktion nur unzureichend erfassen. Insbesondere in Bereichen mit stärkerer Änderung des Funktionsverlaufs sind sichtbare Unterschiede erkennbar. Die Approximation wirkt teilweise zu flach und weicht zwischen den Stützstellen deutlich von der exakten Funktion ab.

```python
plot_spline_approximation(4, -2, 2 * np.pi)
```

### Variante 2: 8 Stützstellen
Mit 8 Stützstellen verbessert sich die Approximation deutlich. Die Abstände zwischen den Punkten sind kleiner, wodurch der Spline den Funktionsverlauf genauer nachbilden kann. Die Abweichungen sind weiterhin vorhanden, jedoch deutlich geringer als bei der ersten Variante. Besonders in den Bereichen mit moderater Krümmung stimmt der Verlauf bereits gut mit der exakten Funktion überein.

```python
plot_spline_approximation(8, -2, 2 * np.pi)
```

### Variante 3: 12 Stützstellen
Bei 12 Stützstellen ist die Approximation nahezu identisch mit der exakten Funktion. Die Unterschiede sind visuell kaum noch erkennbar. Der Spline kann die Krümmung der Funktion über das gesamte Intervall sehr präzise abbilden. Auch die Fehleranalyse zeigt, dass die Abweichungen nur noch minimal sind.

```python
plot_spline_approximation(12, -2, 2 * np.pi)
```

## Diskussion der Ergebnisse

Die Ergebnisse zeigen deutlich, dass die Anzahl und Verteilung der Stützstellen einen entscheidenden Einfluss auf die Qualität der Spline-Approximation haben.

Bei der Verwendung von nur 4 Stützstellen ist die Approximation vergleichsweise ungenau. Die grossen Abstände zwischen den einzelnen Punkten führen dazu, dass der Spline den tatsächlichen Funktionsverlauf nur grob nachbilden kann. Insbesondere in Bereichen mit stärkerer Krümmung, wie beispielsweise im Übergang zwischen lokalen Minima und Maxima der Sinuskomponente, sind deutliche Abweichungen erkennbar. Der Verlauf der Approximation wirkt in diesen Bereichen teilweise zu flach und folgt der exakten Funktion nur eingeschränkt. Dies zeigt, dass eine geringe Anzahl an Stützstellen nicht ausreicht, um komplexere Funktionsverläufe präzise abzubilden.

Mit 8 Stützstellen verbessert sich die Approximation deutlich. Durch die geringeren Abstände zwischen den Punkten kann der Spline den Verlauf der Funktion wesentlich besser erfassen. Die Abweichungen zwischen exakter Funktion und Approximation sind sichtbar reduziert, insbesondere in Bereichen mit moderater Krümmung. Dennoch treten weiterhin kleinere Unterschiede auf, vor allem in den Abschnitten, in denen sich die Krümmung der Funktion stärker verändert. Dies verdeutlicht, dass eine mittlere Anzahl an Stützstellen bereits eine brauchbare, jedoch noch nicht perfekte Approximation ermöglicht.

Bei 12 Stützstellen ist die Approximation nahezu identisch mit der exakten Funktion. Die Unterschiede sind visuell kaum noch wahrnehmbar, und auch die einfache Fehleranalyse zeigt nur noch geringe Abweichungen. Der Spline ist in der Lage, sowohl den linearen Trend als auch die periodische Struktur der Funktion sehr präzise abzubilden. Dies bestätigt, dass mit zunehmender Anzahl an Stützstellen die Approximationsqualität deutlich steigt.

Ein weiterer wichtiger Aspekt ist die Wahl der zugrunde liegenden Funktion. Durch die Kombination aus einer trigonometrischen Komponente und einem linearen Trend ergibt sich ein Funktionsverlauf, der sowohl periodische als auch monotone Eigenschaften aufweist. Dadurch wird die Approximation anspruchsvoller als bei einer reinen Sinusfunktion, da der Spline gleichzeitig unterschiedliche Verhaltensweisen der Funktion erfassen muss. Dies führt dazu, dass Unterschiede zwischen den einzelnen Varianten deutlicher sichtbar werden und die Auswirkungen der Stützstellenanzahl besser analysiert werden können.

Neben der Anzahl der Stützstellen spielt auch deren Verteilung eine zentrale Rolle. In dieser Arbeit wurden die Stützstellen gleichmässig über das gesamte Intervall verteilt. Diese Vorgehensweise ist einfach umzusetzen, jedoch nicht in allen Fällen optimal. Insbesondere in Bereichen mit hoher Krümmung, in denen sich die Funktion schnell verändert, wären zusätzliche Stützstellen sinnvoll, um die Approximation weiter zu verbessern. Umgekehrt könnten in Bereichen mit nahezu linearem Verlauf weniger Stützstellen ausreichend sein. Eine adaptive Platzierung der Stützstellen, bei der die Punktdichte an die Eigenschaften der Funktion angepasst wird, könnte somit zu einer effizienteren und genaueren Approximation führen.

Die Qualität der Approximation wurde in dieser Arbeit sowohl visuell als auch durch eine einfache Fehleranalyse beurteilt. Die grafische Darstellung ermöglicht eine intuitive Einschätzung der Abweichungen, während die Berechnung des maximalen und mittleren Fehlers eine quantitative Bewertung liefert. Beide Ansätze ergänzen sich und tragen zu einem besseren Verständnis der Approximationsgüte bei.

Zusammenfassend lässt sich feststellen, dass kubische Splines eine sehr leistungsfähige Methode zur Approximation nichtlinearer Funktionen darstellen. Bereits mit einer moderaten Anzahl an Stützstellen kann eine hohe Genauigkeit erreicht werden. Gleichzeitig zeigt sich jedoch, dass eine sorgfältige Wahl der Anzahl und Verteilung der Stützstellen entscheidend ist, um eine optimale Approximation zu erzielen.
