# Was ist Analysis?

Analysis ist im Kern die Mathematik der Veränderung.

Während die Algebra fragt: „Was ist das Ergebnis?“, fragt die Analysis:
„Was passiert, wenn etwas immer weiter wächst, schrumpft oder sich annähert?“

Sie untersucht Grenzwerte, Stetigkeit, Ableitungen, Integrale. Alles dreht sich um Prozesse – nicht um statische Zustände.

Der zentrale Gedanke ist der Grenzwert.

Was passiert mit einer Größe, wenn $n$ immer größer wird?
Geht sie gegen 0? Gegen unendlich? Bleibt sie stabil?

Analysis formalisiert genau diese Fragen.

Man kann es so einordnen:
- Folgen → Verhalten diskreter Prozesse
- Funktionen → Verhalten kontinuierlicher Größen
- Ableitungen → Änderungsraten
- Integrale → aufsummierte Wirkung
- Reihen → unendliche Summen als Approximationen

Wenn du „Analysis mit Python“ hörst, bedeutet das meist:

Wir betrachten diese Konzepte nicht nur abstrakt, sondern rechnen, simulieren und visualisieren sie.

Zum Beispiel:

Statt nur zu behaupten, dass

$$
\left(1 + \frac{1}{n}\right)^n \to e
$$

lässt du Python $n$ immer größer werden und beobachtest numerisch die Annäherung.

Oder du approximierst eine Funktion durch eine Reihe und siehst grafisch, wie die Näherung besser wird.

Python wird hier zum Experimentierlabor für Grenzprozesse.
Man beobachtet numerische Stabilität, Rundungsfehler, Konvergenzgeschwindigkeit. Das sind reale Effekte, keine Schulbuchtheorie.

Analysis ist also:
- die Theorie des Unendlichen in endlichen Schritten
- die Sprache für Wachstum und Approximation
- das Fundament für Optimierung, Wahrscheinlichkeit und numerische Verfahren

Mit Python bekommt das Ganze eine zweite Ebene:
Du siehst nicht nur Beweise, sondern Verhalten.

# Vollständige Induktion aus Teschl (Kapitel 2.3)
Es ist schwer, eine Summe mit variablen Grenzen zu berechnen. Zum Beispiel: Was ist die Summe der ersten n natürlichen Zahlen?

$1 + 2 + 3 + ... + n = ?$

Manchmal ist es mögilch, eine Formel für die Summe zu erraten. In diesem Fall könnte man die Formel $S(n) = \frac{n(n+1)}{2}$ erraten. Aber wie kann man sicher sein, dass diese Formel korrekt ist? Hier kommt die vollständige Induktion ins Spiel. Es ist eine Methode, um zu beweisen, dass eine Aussage für alle natürlichen Zahlen gilt.

Für das obige Beispiel kann man kleine Zahlen zur Überprüfung verwenden:
- Für n=1: $S(1) = 1$ und $\frac{1(1+1)}{2} = 1$ (korrekt)
- Für n=2: $S(2) = 1 + 2 = 3$ und $\frac{2(2+1)}{2} = 3$ (korrekt)
- Für n=3: $S(3) = 1 + 2 + 3 = 6$ und $\frac{3(3+1)}{2} = 6$ (korrekt)

## Satz 2.27 (Induktionsprinzip oder Vollständige Induktion): Sei A(n) eine Aussage für beliebiges n ∈ ℕ, sodass gilt:
- Induktionsanfang: A(1) ist wahr und
- Induktionsschluss: Aus der Richtigkeit von A(n) für ein beliebiges, festes n ∈ ℕ (Induktionsvoraussetzung) folgt die Richtigkeit von A(n + 1). (Anstelle der Richtigkeit von A(n) kann sogar die Richtigkeit von A(k) für alle k <= n vorausgesetzt werden.)

Darum ist A(n) für alle n ∈ ℕ richtig.

    Das Induktionsprinzip ist wie der Dominoeffekt. Sie m¨ochten, dass alle Steine umfallen (dass die
    Aussage fur alle ¨ n bewiesen wird). Dazu mussen Sie den ersten Stein anstoßen (Induktionsanfang) ¨
    und es muss sichergestellt sein, dass ein beliebiger Stein den darauf folgenden umwirft (Schluss von
    n auf n + 1).

Die Induktion muss nicht bei 1 beginnen, sondern kann auch angewendet werden, wenn eine Aussage für alle ganze Zahlen ab einer bestimmten Zahl $n_0$ ∈ $Z$ (z.B. $n_0 = 0$ oder $n_0 = 2$) formuliert wird.

## Deuten der Formel-Darstellung
Wie deutet man die folgende Formel?

$$
\sum_{j=1}^{n} (2j - 1) = ?
$$

Formuliert: Summiere den Ausdruck 2j-1, wobei j bei 1 beginnt und bis n läuft.

- Das grosse Sigma (∑) bedeutet: addiere mehrere Terme hintereinander
- Unten steht j = 1. Das ist der Startwert
- Oben steht n. Das ist der Endwert
- Der Ausdruck rechts davon, also 2j - 1, sagt dir, was du bei jedem Schritt einsetzen sollst

Vorgehen ist also so:
- Für j=1: 2*1 -1 = 1
- Für j=2: 2*2 -1 = 3
- Für j=3: 2*3 -1 = 5
- ...
- Für j=n: 2*n -1

Hier werden also alle ungeraden Zahlen von 1 bis zur n-ten ungeraden Zahl addiert. Die Summe sieht ausgeschrieben so aus:
$$
1 + 3 + 5 + ... + (2n - 1) = ?
$$

Die Summe ergibt hier also immer $n^2$. Das ist kein Zufall. Die Summe der ersten n ungeraden Zahlen ist ein perfektes Quadrat.

Strukturell passiert also Folgendes:
- j ist nur ein Laufindex (eine Zählvariable)
- 2j - 1 erzeugt systematisch die ungeraden Zahlen
- Das Sigma addiert sie auf
- Das Resultat ist $n^2$

Summenzeichen sind somit einfach nur kompakte For-Schleifen in mathematischer Notation.

# Kapitel 6 – Folgen und Reihen

---

## 6.1 Folgen

### Grundidee

Eine **Folge** ist eine Funktion
$$
a : \mathbb{N} \to \mathbb{R}, \quad n \mapsto a_n
$$

Man schreibt sie als

$$
a_1, a_2, a_3, \dots
$$

- $a_n$ heißt **n-tes Folgenglied**
- $n$ heißt **Index**
- Das Bildungsgesetz gibt an, wie $a_n$ berechnet wird

Der Index muss nicht bei 1 beginnen, er kann auch bei 0 starten.

---

### Explizite und rekursive Definition

**Explizit:**
$$
a_n = n^2
$$

Hier kann jedes Glied direkt berechnet werden.

**Rekursiv:**
$$
a_1 = 1, \quad a_n = n \cdot a_{n-1}
$$

Hier wird jedes Glied aus dem vorherigen berechnet.

Beispiel:
$$
a_1 = 1,\quad a_2 = 2,\quad a_3 = 6,\quad a_4 = 24,\dots
$$

Das ist die Fakultätsfolge $a_n = n!$.

---

### Monotonie und Beschränktheit

Eine Folge heißt:

- **monoton wachsend**, wenn $a_n \le a_{n+1}$
- **monoton fallend**, wenn $a_n \ge a_{n+1}$
- **beschränkt**, wenn es Zahlen $k, K$ gibt mit
  $$
  k \le a_n \le K
  $$

Wichtig:
> Jede konvergente Folge ist beschränkt.

---

### Konvergenz

Eine Folge konvergiert gegen $a$, wenn gilt:

$$
\forall \varepsilon > 0 \ \exists n_0 \ \text{sodass} \ |a_n - a| < \varepsilon \text{ für alle } n \ge n_0
$$

Schreibweise:

$$
\lim_{n \to \infty} a_n = a
$$

Das bedeutet:
Die Folgenglieder kommen dem Wert $a$ beliebig nahe.

---

### Wichtige Grenzwerte

**Fundamentale Nullfolgen:**

$$
\lim_{n \to \infty} \frac{1}{n} = 0
$$

$$
\lim_{n \to \infty} q^n = 0 \quad \text{für } |q| < 1
$$

---

### Divergenz

Eine Folge ist **divergent**, wenn sie keinen Grenzwert besitzt.

Beispiele:

- $a_n = (-1)^n$ (oszilliert)
- $a_n = 2^n$ (wächst unbeschränkt)

---

### Bestimmte Divergenz

Man schreibt:

$$
\lim_{n \to \infty} a_n = \infty
$$

wenn:

- die Folge ab einem Index positiv ist
- die Folge unbeschränkt wächst

Analog für $ -\infty $.

---

### Rechenregeln für konvergente Folgen

Sind $a_n \to a$ und $b_n \to b$, dann gilt:

$$
\lim (a_n \pm b_n) = a \pm b
$$

$$
\lim (a_n b_n) = ab
$$

$$
\lim \frac{a_n}{b_n} = \frac{a}{b}, \quad b \ne 0
$$

---

### Rationale Folgen

Bei Brüchen von Polynomen entscheidet der Grad:

$$
\frac{a_k n^k + \dots}{b_\ell n^\ell + \dots}
$$

- $k < \ell$ → Grenzwert 0
- $k = \ell$ → Grenzwert = Quotient der Leitkoeffizienten
- $k > \ell$ → ±∞

---

### Konvergenzkriterien für Folgen

1. Jede monotone und beschränkte Folge konvergiert.
2. Produkt aus beschränkter Folge und Nullfolge ist Nullfolge.

---

### Euler’sche Zahl

$$
e = \lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n
$$

---

### Heron-Verfahren (Wurzelberechnung)

Rekursion:

$$
h_n = \frac{1}{2} \left(h_{n-1} + \frac{x}{h_{n-1}}\right)
$$

Konvergiert für jeden Startwert $h_1 > 0$ gegen:

$$
\sqrt{x}
$$

---

## 6.2 Reihen

### Definition

Eine **Reihe** ist der Ausdruck:

$$
\sum_{k=0}^{\infty} a_k
$$

Definiert wird sie über die **Teilsummenfolge**:

$$
s_n = \sum_{k=0}^{n} a_k
$$

Konvergiert $s_n$, dann konvergiert die Reihe.

---

### Notwendiges Kriterium

Wenn eine Reihe konvergiert, dann gilt:

$$
a_k \to 0
$$

Achtung:
Die Umkehrung gilt nicht (harmonische Reihe!).

---

### Geometrische Reihe

$$
\sum_{k=0}^{\infty} q^k
$$

Teilsumme:

$$
s_n = \frac{1 - q^{n+1}}{1 - q}
$$

Grenzwert:

$$
\sum_{k=0}^{\infty} q^k = \frac{1}{1-q}, \quad |q| < 1
$$

Für $ |q| \ge 1 $ divergent.

---

### Wichtige Beispiele

$$
\sum \frac{1}{k} \quad \text{divergent}
$$

$$
\sum \frac{1}{k^2} = \frac{\pi^2}{6}
$$

---

### Majorantenkriterium

Ist

$$
|a_k| \le b_k
$$

und konvergiert $\sum b_k$,
dann konvergiert auch $\sum a_k$.

---

### Quotientenkriterium

Ist

$$
\lim_{k \to \infty} \left|\frac{a_{k+1}}{a_k}\right| = q
$$

- $q < 1$ → Reihe konvergent
- $q > 1$ → divergent
- $q = 1$ → keine Aussage

---

### Potenzreihen

Beispiele:

$$
\sum_{k=0}^{\infty} \frac{x^k}{k!} = e^x
$$

$$
\sum_{k=1}^{\infty} \frac{x^k}{k} = -\ln(1-x), \quad |x| < 1
$$

---

## 6.3 Rechnen mit dem Computer

### Folgen berechnen

Beispiel (symbolisch gedacht):

$$
a_n = \frac{1}{2^n}
$$

Grenzwerte:

$$
\lim_{n \to \infty} a_n
$$

---

### Reihen berechnen

$$
\sum_{k=0}^{\infty} \frac{1}{2^k} = 2
$$

Teilsummen approximieren Grenzwerte numerisch.

---

# Kerngedanken des Kapitels

- Folgen modellieren Grenzprozesse.
- Konvergenz bedeutet: beliebig genaue Annäherung.
- Reihen sind Grenzwerte von Summen.
- Geometrische Reihen sind die zentrale Referenz.
- Monotonie + Beschränktheit → Konvergenz.
- Nullfolge ist notwendige Bedingung für Reihen.
- Viele Grenzwerte beruhen auf Vergleichs- oder Quotientenkriterien.
