# Vorbemerkung bzw. vollständige Induktion
Peamo-Axiome:
- (A1): 1 ist eine natürliche Zahl (1 ∈ ℕ)
- (A2): Falls M ∈ ℕ => M + 1 ∈ ℕ

Jede natürliche Zahl hat einen natürlichen Nachvolger.

Die Peamo-Axiome bilden die Grundlage für die vollständige Induktion, ein Beweisverfahren, das verwendet wird, um Aussagen über alle natürlichen Zahlen zu beweisen. Es besteht aus zwei Schritten:
1. Basisfall: Zeige (rechne aus), dass die Aussage für die kleinste natürliche Zahl (normalerweise 1 oder 0) wahr ist.
2. Induktionsschritt: Zeige (rechne aus), dass wenn die Aussage für eine beliebige natürliche Zahl n wahr ist, sie auch für n + 1 wahr sein muss.
3. Wenn beide Schritte erfolgreich durchgeführt wurden, kann man schließen, dass die Aussage für alle natürlichen Zahlen wahr ist.

## Formale Aussage
Sei $m_0 \in ℕ$ und A(m) für $m \geq m_0$ eine Ausageform.
Es genügt zu zeigen:
1. A($m_0$) ist wahr (Basisfall/Induktionsanfang) => Induktionsanfangen (IA)
2. Wir nehmen an, dass A(k) richtig ist $k \geq m$ => Induktionsvoraussetzung (IV)
3. Falls IV stimmt und IV eingeht, dann m --> m + 1 => Induktionsschritt (IS)

## Beispiele
1) $x^m - 2 = 0$ => keine Induktion, da Gleichung gelöst werden muss
2) $$\sum_{k=1}^{m} k = 1 + 2 + ... + m = \frac{m (m + 1)}{2}$$ => Induktion möglich, da Aussageform über m
3) $$\prod_{k=1}^{m} k = 1 * 2 * ... * m = m!$$ => Induktion möglich, da Aussageform über m
4) ???

## Konkretes Beispiel
Wir beweisen $$\sum_{k=1}^{m} k = 1 + 2 + ... + m = \frac{m (m + 1)}{2}$$

IA: m = 1 (nicht 0, weil die Formel bei 1 (k) beginnt)
$$\sum_{k=1}^{1} k = 1 = \frac{1 (1 + 1)}{2}$$ => IA ist wahr
Das stimmt, weil die Formel rechts 1 ergibt. Und die Summe links auch 1 ergibt.

IV: $$\sum_{k=1}^{m} k = \frac{m (m + 1)}{2}$$   gelte, k <= m

IV mus nun korrekt eingebaut werden:

IS: m -> m + 1
$$\sum_{k=1}^{m + 1} k = \sum_{k=1}^{m} k + (m + 1)$$ => der rechte Teil ist die Umformulierung des linken Teils des Gleichzeichens

=> $$\sum_{k=1}^{m} k + (m + 1) = \frac{m(m+1)}{2} + m+1$$

=> $$\frac{m(m+1)}{2} + \frac{2(m+1)}{2} = \frac{m(m+1) + 2(m+1)}{2} = \frac{m^2 + 3m + 2}{2} = \frac{(m+1)(m+2)}{2}$$

Wichtig ist, dass die Induktionsvoraussetzung (IV) korrekt eingebaut wird. Sonst ist der Beweis automatisch falsch.



## Beispiel 2
Zu zeigen: $$\prod_{k=1}^{m} 4^k = 2^{m(m+1)}$$

IA: m = 1
$$\prod_{k=1}^{1} 4^k = 4^1 = 4 == 2^{1(1+1)} = 2^2 = 4$$ => IA ist wahr

IV: $$\prod_{k=1}^{m} 4^k = 2^{m(m+1)}$$   gelte, k <= m

IS: $$\prod_{k=1}^{m + 1} 4^k = \prod_{k=1}^{m} 4^k * 4^{m + 1}$$ => der rechte Teil ist die Umformulierung des linken Teils des Gleichzeichens

=> $$2^m(m+1) * 4^{m + 1} = 2^{m^2+m} * (2^2)^{m + 1} = 2^{m^2 +m} * 2^{2m + 2} = 2^{M2 + m + 2m + 2} = 2^{m^2 + 3m + 2} = 2^{(m + 1)(m + 2)}$$

## Übung 1

Beweise mit Induktion:
$$\sum_{j=1}^{m} (2j-1) = m^2$$

IA: m = 1
$$\sum_{j=1}^{1} (2j-1) = 2 * 1 - 1 = 1 == 1^2$$

IV: $$\sum_{j=1}^{m} (2j-1) = m^2$$   gelte, j <= m

IS: $$\sum_{j=1}^{m + 1} (2j-1) = \sum_{j=1}^{m} (2j-1) + 2(m+1) - 1$$ => der rechte Teil ist die Umformulierung des linken Teils des Gleichzeichens

=> = $$m^2 + 2m + 1 = (m+1)^2$$



# Ungleichungen
Hier kann es vorkommen, dass eine Induktion nicht ausreicht.

Beispiel:

Beweise $2^m > m$   gelte, m >= 1   | Hier muss eigentlich bewiesen werden, dass 2^{m+1} grösser ist als m + 1

Vermutung: 2^1 > 1 => IA ist wahr
Vermutung: 2^2 > 2 => IA ist wahr
Vermutung: 2^3 > 3 => IA ist wahr

Sieht also richtig aus.

IA: m = 1
$$2^1 > 1$$ => IA ist wahr

IV:
$$ 2^m > m$$ ist korrekt  | Das muss jetzt im Induktionsschritt eingebaut werden

IS:
$$2^{m + 1} = 2 * 2^m$$  | Das ist die Umformulierung des linken Teils des Gleichzeichens

Neue Gleichung: $$2 * 2^m > 2 * m$$

Die beiden 2-Mal können gestrichen werden, damit die Gleichung einfacher wird:
$$2^m > m$$

# Folgen
Es gibt Folgen und Rekursiv definierte Folgen (Rekursionen).

Folge: $a_m: ℕ \to ℝ$   | Das bedeutet, wir ordnen jeder $m \in ℕ$ eine reelle Zahl zu.

$a_m = (a_m) = (a_m)_m = (a_1, a_2, a_3, ...)$

$a_m = \frac{1}{m} \to a_m = (1, \frac{1}{2}, \frac{1}{3}, ...)$

Den Teil vor dem Pfeil nennt man Explizite Folge. Der restliche Teil ist die Rekursion.

---

Alternierend: bedeutet, dass sich beispielsweise das Vorzeichen (+ oder -) abwechselt.

---

Bei einer expliziten Folge erhalten wir das Resultat direkt durch Einsetzen der Zahl m.

Bei einer rekursiven Folge benötigen wir die vorherigen Werte, um den nächsten Wert zu berechnen.

---

Umwandlung von rekursiv definierter Folge in explizite Form:
1. Berechne die ersten paar Werte der rekursiven Folge, um ein Muster zu erkennen.
2. Versuche, eine allgemeine Formel zu finden, die die Werte der Folge beschreibt.
3. Überprüfe die gefundene Formel, indem du sie in die rekursive Definition einsetzt, um sicherzustellen, dass sie korrekt ist.

Beispiel:<br>
$a_0 = 2, a_m = 2 - \frac{1}{a_{m-1}}$ gilt für alle $m \geq 1 $

Ein paar Werte berechnen:<br>
$a_1 = 2 - \frac{1}{a_0} = 2 - \frac{1}{2} = \frac{3}{2}$<br>
$a_2 = 2 - \frac{1}{a_1} = 2 - \frac{1}{\frac{3}{2}} = \frac{4}{3}$<br>
$a_3 = 2 - \frac{1}{a_2} = 2 - \frac{1}{\frac{4}{3}} = \frac{5}{4}$

Das Muster ist erkennbar: $a_m = \frac{m + 2}{m + 1}$

Beweis:<br>
$a_m = \frac{m + 2}{m + 1}$  | zu zeigen ist $a_{m+1} = \frac{m+3}{m+2}$<br>
